```python
import numpy as np
import math
import sys
```

```python
class learnable_line:
    def __init__(self, learning_rate):
        self.lr = learning_rate
        self.a = np.random.random()
        self.b = np.random.random()
        
    def evaluate(self, x):
        return x * self.a + np.ones_like(x) * self.b
    
    def fit(self, x, y):
        y_pred = self.evaluate(x)
        n_points = len(x)
        loss = np.sum(np.power((y_pred - y), 2.0)) / n_points
        ga = -2 * np.add.reduce((y_pred - y) * x) / n_points
        gb = -2 * np.add.reduce((y_pred - y)) / n_points
        self.a = self.a + self.lr * ga 
        self.b = self.b + self.lr * gb
        return loss
        
    def show(self):
        print "a = ", self.a
        print "b = ", self.b
```

```python
def train(training_x, training_y, lr, epochs, batch_size):
    model = learnable_line(learning_rate=lr)
    for k in range(epochs):
        for s in range(0, len(training_x), batch_size):
            x = training_x[s:s+batch_size]
            y = training_y[s:s+batch_size]
            loss = model.fit(x, y)
            if k%1000 == 0:
                print "loss = ", loss
                sys.stdout.flush()
    return model
```

```python
EPOCHS=10000
LEARNING_RATE=0.00001
BATCH_SIZE = 100
x = np.linspace(-50, 50, 5000)
noise = np.random.random((len(x))) * 0.01 - 0.005
y = x*2 + np.ones_like(x)*3 + noise
model = train(x, y, lr=LEARNING_RATE, epochs=EPOCHS, batch_size=BATCH_SIZE)
model.show()
```
